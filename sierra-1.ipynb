```python
url = "https://turismoactiva.com/rutas-de-montana-sierra-de-gredos/"

res_ruta = requests.get(url)
if res_ruta.status_code == 200:
    print(res_ruta.status_code)
else:
    print("None")
```

```python
sopa_rutas = BeautifulSoup(res_ruta.content, "html.parser")
#print(sopa_rutas.prettify())
```

```python
#Cogemos el nombre de la ruta
nombres_rutas = sopa_rutas.find_all("h4", {"class": "elementor-heading-title elementor-size-default"})
lista_nombres = []
for i in nombres_rutas:
    lista_nombres.append(i.getText())
lista_nombres[:13]
```

```python
datos_rutas = sopa_rutas.find_all("div", {"class": "elementor-widget-container"})
lista_datos = []
for i in datos_rutas:
    lista_datos.append(i.getText())
lista_datos[15:54]
```

```python
datos_rutas = sopa_rutas.find_all("div", {"class": "elementor-widget-container"})
diccionario_datos = {}
for index, elemento in enumerate(datos_rutas):
    diccionario_datos[f"elemento_{index}"] = elemento.getText()
```

```python
for key, value in diccionario_datos.items():
    if int(key.split('_')[1]) >= 15 and int(key.split('_')[1]) <= 54:
        print(value)
```
